```python
import numpy as np
marks_array = np.array([[23,54,67],
                  [87,98,87],
                  [98,88,78],
                  [56,87,45],
                  [56,89,65]])
print(marks_array)
```

```python
print(marks_array.shape)
```

```python
print(marks_array.ndim)
```

```python
print("Average Marks:", np.mean(marks_array))
print("Max Marks:" , np.max(marks_array))
print("Min Marks:" , np.min(marks_array))
print("Sum of Marks:" , np.sum(marks_array))
```

```python
#column wise subject--axis=0 and row wise subject--axis=1
```

```python
subject_avg = np.mean(marks_array, axis=0)
subject_max = np.max(marks_array, axis=0)
subject_min = np.min(marks_array, axis=1)
print(subject_avg)
print(subject_max)
print(subject_min)
```

```python
#count pass and fail student in column and row wise
```

```python
student_pass = np.sum(marks_array >=50, axis=1)
student_fail = np.sum(marks_array <=30, axis=1)

print(student_pass)
print(student_fail)
```

```python
student_pass = np.sum(marks_array >=50, axis=0)
student_fail = np.sum(marks_array <=30, axis=0)

print(student_pass)
print(student_fail)
```

```python
# find toppers
```

```python
topper_student = np.sum(marks_array >=90, axis=1)
print(topper_student)
```

```python
toppr_index = np.argmax(topper_student)
print(topper_index)
```

```python
# student grades
```

```python
grade = np.where(marks_array>=90, 'A',
    np.where(marks_array>=80,'B',
        np.where(marks_array>=50,'C',
            np.where(marks_array>=30,'D',
                np.where(marks_array>=20, 'E','F')
            ))
        ))


print(grade)
```

```python
for avg in student_avg:
  if avg >= 90:
    grade.append('A')
    elif avg >= 80:
      grade.append('B')
      elif avg >= 50:
        grade.append('C')
        elif avg >=30:
          grade.append('D')
          elif avg >=20:
            grade.append('E')
            else:
            grade.append('F')
```

```python
print("Grade",grade)
```
